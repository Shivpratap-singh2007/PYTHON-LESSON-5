### <u> Loops in python </u>   
1.Loops are used to repeat instructions.    

while Loops

    while condition:
                #some work   
       

```python
count = 1 
while count <= 5 :     
    print("KARMA")
    count += 1
print(count)    
```

```python
i = 1 
while i <= 5 :
    print(i)
    i += 1
```

### <u> Brack & Conntinue </u>
Brack : used to terminate the loop when encountered  


Continue : terminates execution in the current iteration & continues execution of the loop with the next iteration.  

```python
i = 1 
while i <= 5 :
    print(i)
    if(i == 3) :
        break
    i += 1
print("Loop ended")    
```

```python
i = 0 
while i <= 5 :
    i += 1
    if(i == 3) :
        i += 1
        continue
    print(i)
    
```

```python
i = 0 
while i <= 5 :
    i += 1
    if(i == 3) :
        i += 1
        continue
    print(i)
    i += 1
```

2. Loops are used for sequential traversal. for traversing list,string, tuples etc.        

     For Loops     
 for el in list:

    For loops with else     
  for el in list:    
     #some work   
  else:   
    #work when loop ends     
              

```python
list =  [1, 2, 3, 4, 5] 
for el in list :
    print(el) 
    
```

```python
list =  [1, 2, 3, 4, 5] 
for el in list :
    print(el) 
else: 
    print("END")    
```

### <u> Range() </u>     
Range functions returns a sequence of numbers, stsrting from 0 by default, and increments by 1 (by default), and stops before a specified number.          

range(start?,stop,step?)
      

```python
seq = range(5)

for el in seq : 
    print(el)
```

```python
for i in range(10):     #range(stop)
    print(i)
```

```python
for i in range(2,10,2):        #range(start, stop, step)
    print(i)
```
